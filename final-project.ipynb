```python
import requests
from bs4 import BeautifulSoup
```

```python
i=[1,2,3,4,5,6,7,8]
for i in range(1):
    url="https://www.zameen.com/Houses_Property/Karachi_Malir_Model_Colony___Malir-518-1.html".format(i)

    r = requests.get(url)
```

```python
r
```

```python
soup = BeautifulSoup(r.text, 'html.parser') 
```

```python
result = soup.find_all('article')
```

```python
len(result)
```

```python
records=[]
for i in range (len(result)):
    PRICEFLAT=soup.find_all('span', attrs={'aria-label': 'Listing price'})
    CURRENCYFLAT=soup.find_all('span', attrs={'aria-label': 'Listing currency'})
    LOCATIONFLAT=soup.find_all('div', attrs={'aria-label': 'Listing location'})
    BEDSFLAT=soup.find_all('span', attrs={'aria-label': 'Beds'})
    BATHSFLAT=soup.find_all('span', attrs={'aria-label': 'Baths'})
    AREAFLAT=soup.find_all('span', attrs={'aria-label': 'Area'})
    PURPOSE=soup.find('div', attrs={'class': 'e7c6503c',})
    PURPOSE=PURPOSE.text
    PROPERTYTYPE=soup.find('div', attrs={'aria-label': 'Category filter'})
    TYPE=PROPERTYTYPE.text
    BATHS=BATHSFLAT[i].text
    BEDS=BEDSFLAT[i].text
    LOCATION=LOCATIONFLAT[i].text
    CURRENCY=CURRENCYFLAT[i].text
    PRICE=PRICEFLAT[i].text
    AREA=AREAFLAT[i].text
    
    records.append((TYPE,PURPOSE,CURRENCY,PRICE,LOCATION,BEDS,BATHS,AREA))
    
```

```python
len(records)
```

```python
import numpy as np
import pandas as pd 
from pandas import DataFrame
```

```python
df = pd.DataFrame(records, columns=['TYPE','PURPOSE','CURRENCY', 'PRICE', 'LOCATION', 'BEDS','BATHS','AREA'])

display(df)

```

```python
df.to_csv("Clean.csv", header=True, index=True)
```

```python
s=[];
d=[];
c=[];
for x in range(len(df.AREA)):    

    s=re.findall('[-+]?\d*\.\d+|\d+', df.AREA[x])
    d= df.AREA[x].split()
    #print(d);
for i in range(len(df.AREA)):
    d= df.AREA[i].split()
    #print(d);
    if(d[1]=='Kanal'):
        c.append(float(re.sub("[^\d\.]", "", d[0]))*605)
    elif(d[1]=='Marla'): 
        c.append(float(re.sub("[^\d\.]", "", d[0]))*31)
    else:
        c.append(float(re.sub("[^\d\.]", "", d[0])))
    

    
    
df['AREA_Sqyd']=c;
print(c);
```

```python
df.head()
```

```python

s=[];
d=[];
c=[];
m=[];
for x in range(len(df.PRICE)):    

    s=re.findall('[-+]?\d*\.\d+|\d+', df.PRICE[x])
    d= df.PRICE[x].split()
    #print(d);


for i in range(len(df.PRICE)):
    d= df.PRICE[i].split()
    m=d[-1];
    if(m=='Crore'):
        c.append(float(re.sub("[^\d\.]", "", d[0]))*10000000)
    elif(m=='Lakh'): 
        c.append(float(re.sub("[^\d\.]", "", d[0]))*100000)
    elif(m=='Thousand'):
        c.append(float(re.sub("[^\d\.]", "", d[0]))*1000)
    else:
        c.append(0)
            
        
print(len(c))
df['Price_New']=c;
print(c);
```
